```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {
    'Date': pd.date_range(start='2023-01-01', periods=30, freq='D'),
    'Category': np.random.choice(['A', 'B', 'C', 'D'], size=30),
    'Sales': np.random.randint(100, 1000, size=30),
    'Profit': np.random.uniform(10, 100, size=30),
    'Region': np.random.choice(['North', 'South', 'East', 'West'], size=30),
    'Customer_Age': np.random.randint(18, 65, size=30)
}

df = pd.DataFrame(data)
df
```

# Bar plot

```python
S=df.groupby("Category")["Sales"].sum().reset_index()
S
```

```python
colors=["green", "red", "blue", "black"]
categories=["c.A", "c.B", "c.C", "c.D"]
plt.bar(S["Category"], S["Sales"], color=colors, alpha=0.8, label=categories, tick_label=categories)
plt.xlabel("Categories")
plt.ylabel("Sales")
plt.title("Sales by Categories")
plt.legend()
plt.show()
```

# Line plot

```python
plt.plot(df["Date"], df["Sales"],marker="o", label="Sales")
plt.plot(df["Date"], df["Profit"], label="Profit")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.legend(loc=('upper right'))
plt.show()
```

# Histogram

```python
plt.hist(data["Customer_Age"], bins=10, edgecolor='black', linewidth=1.2)
plt.xlabel("Customer Age")
plt.ylabel("Number of Customers")
plt.title("Distribution of Customer Age")
plt.show()
```

# Scatter plot

```python
plt.scatter(x=df["Sales"], y=df["Profit"], color="red", s=40, alpha=0.8)
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.title("Sales Vs Profit")
plt.show()
```

# Pie plot

```python
sal_avg=df.groupby("Region")["Sales"].mean().reset_index()
sal_avg
```

```python
plt.pie(sal_avg["Sales"], autopct='%1.2f%%', labels=sal_avg["Region"], explode=(0,0,0,0.05))
plt.title("Percentage of Sales by Region")
plt.show()
```
